In this challenge : analyze demographic data using Pandas. You are given a dataset of demographic data that was extracted from the 1994 Census database.

Reading the data file, analizing basic columns, shape of the data

```python
import numpy as np
import pandas as pd 

df= pd.read_csv("Census_1994.csv")
print(df.describe())
df.shape
df.head(20)

```

How many of each race are represented in this dataset? This should be a Pandas series with race names as the index labels.

```python
count_race = pd.Series(df['race'].value_counts()).values
print (count_race)
```

What is the average age of men?

```python
Avg_age_men = np.mean(df['age'].where (df['sex'] == 'Male' ))
print(Avg_age_men)
```

What is the percentage of people who have a Bachelor's degree?

```python
ppl_with_Bachelor = df.loc[df['education']=='Bachelors', 'education'].count()
total_ppl = len(df)

percentage_Bachelors = round((ppl_with_Bachelor/total_ppl)*100,3)
percentage_Bachelors
```

What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?
What percentage of people without advanced education make more than 50K?

with and without `Bachelors`, `Masters`, or `Doctorate`

```python
advance_edu = ['Bachelors', 'Masters', 'Doctorate'] 
higher_education = df.loc[df['education'].isin(advance_edu), 'education'].count()
lower_education = df.loc[~df['education'].isin(advance_edu), 'education'].count()
higher_education_rich = round(df.loc[(df['education'].isin(advance_edu)) & (df['salary'] == '>50K'), 'education'].count() / higher_education * 100, 1)
lower_education_rich = round(df.loc[(~df['education'].isin(advance_edu)) & (df['salary'] == '>50K'), 'education'].count() / lower_education * 100, 1)
higher_education_rich
```

What is the minimum number of hours a person works per week?
What percentage of the people who work the minimum number of hours per week have a salary of more than 50K?

```python
min_work_hours = df['hours-per-week'].min()
print(f"Minimum hours worked by ppl: {min_work_hours}")
num_min_workers = df.loc[df['hours-per-week'] == min_work_hours, 'hours-per-week'].count()
rich_percentage = round((df.loc[(df['hours-per-week'] == min_work_hours) & (df['salary'] == '50K'), 'hours-per-week'].count() / num_min_workers) * 100, 1)
rich_percentage
```

What country has the highest percentage of people that earn >50K and what is that percentage?

```python
highest_earning_country = df.loc[df['salary'] == '>50K', 'native-country'].value_counts().idxmax()
highest_earning = df.loc[df['salary'] == '>50K', 'native-country'].value_counts().max()
highest_earning_country_percentage = round((highest_earning / total_ppl) * 100,1)
highest_earning_country
highest_earning_country_percentage
highest_earning_country
```

Identify the most popular occupation for those who earn >50K in India.

```python
top_IN_occupation = df['occupation'].loc[(df['native-country'] == 'India') & (df['salary'] == '>50K')].value_counts().idxmax()
top_IN_occupation
```
